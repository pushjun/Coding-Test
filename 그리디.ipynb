```python
x,y= (1,1)
```

```python
n
```

```python
dx = [1,-1,0,0] #+,-

dy = [0,0,-1,1] #  -1,  1
```

```python
while n == 0:
    for i in control:
        if i == 'R':
            x+=dx[0]
            n-=1
            if x==0:
                x+=1
        elif i == 'L':
            x+=dx[1]
            n-=1
            if x==0:
                x+=1  
            
        elif i == 'U':
            y+=dy[2]
            n-=1
            if y==0:
                y+=1 
                
        elif i == 'D':
            y+=dy[3]
            n-=1
            if y==0:
                y+=1 
                
print(x,y)
```
